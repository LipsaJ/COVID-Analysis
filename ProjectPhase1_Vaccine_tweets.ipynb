#### Collect Vaccine related tweets

Import required packages

```python
import tweepy
import json
import csv
import pandas as pd
import os
```

Twitter data collection set up

```python
# Function to read the key file and load keys in a dictionary
def loadKeys(key_file):
    with open(key_file) as f:
        key_dict = json.load(f)
    return key_dict['api_key'], key_dict['api_secret'], key_dict['token'], key_dict['token_secret']

KEY_FILE = 'keys.json'
api_key, api_secret, token, token_secret = loadKeys(KEY_FILE)
auth = tweepy.OAuthHandler(api_key, api_secret)
auth.set_access_token(token, token_secret)
api = tweepy.API(auth)
```

```python
file = 'vaccinetweets.csv'
if(os.path.exists(file) and os.path.isfile(file)): 
  os.remove(file) 
  print("file deleted") 
else: 
  print("file not found") 
```

```python
csvFile = open('vaccinetweets.csv', 'a')
#Use csv writer
csvWriter = csv.writer(csvFile)
```

Saving the tweets in the CSV file

```python
search_term = "Vaccine OR vaccine against coronavirus OR Vaccines work OR Covid Vaccine OR #vaccine2021"

new_search = search_term 

for page in tweepy.Cursor(api.search, wait_on_rate_limit=True, q = new_search, lang="en",tweet_mode="extended",since="2021-03-02").pages():
    for status in page:
        csvWriter.writerow([status.id ,status.user.name , status.full_text,status.entities , status.retweet_count, status.favorite_count, status.created_at])
        
            
```
